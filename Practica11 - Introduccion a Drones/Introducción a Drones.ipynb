**2. Realizar un programa en python para conectarse con el drone y acceder a la camara del mismo.**

Crea el codigo necesario para conectarce al drone, imprime en consola la bateria del mismo y conectate a la camara del drone y reescala para que sea de 360x248 y muestra la imagen al precionar la tecla q se debe cerrar la vista de la camara, realiza la conexión del drone y ejecute el codigo.

```python
from djitellopy import tello
import time
import cv2 as cv

drone = tello.Tello()
drone.connect()
print(drone.get_battery())

drone.streamon()
while True:
    img = drone.get_frame_read().frame
    img = cv.resize(img, (360,248))
    cv.imshow('Drone', img)
    k = cv.waitKey(1)
    if k == ord('q'): break

cv.destroyAllWindows()
```

**3. Realizar un programa en Phyton con el cual se pueda mover el drone desde la computadora haciendo que despeque se mueva y aterrice.**

Complete el codigo faltante, realiza la conexión del drone y ejecute el codigo.

```python
from djitellopy import Tello
import time
import cv2 as cv

width = 360
height = 248
startCounter = 0

#Conneccion al drone
me = Tello()
me.connect()
me.for_back_velocity = 0
me.left_rigth_velocity = 0
me.up_down_velocity = 0
me.yaw_velocity = 0
me.speed = 0

#Mustra la bateria en consola
print(me.get_battery())

drone.streamoff()
drone.streamon()

while True:
  #Obten la imagen de la camara
  frame_read = me.get_frame_read()
  myFrame = frame_read.frame

  #Rescala la iamgen con los valores de width y height
  img = cv.resize(myFrame, (width, height))

  #Cree un condicional donde el drone rote 90 grados, se mueva a la derecha y luego a la izquierda por cada movimiento dar una espera de 3 segundos al final aterrizarlo
  if startCounter == 0:
    me.takeoff()
    time.sleep(8)
    me.rotate_clockwise(90)
    time.sleep(3)
    me.move_left(35)
    time.sleep(3)
    me.land()
    startCounter = 1

  #Muestra la imagen de lo que ve el drone
  cv.imshow('Drone', img)

  #Al presionar tecla q aterrice el drone
  k = cv.waitKey(1)
  if k == ord('q'): 
    me.land()
    break

cv.destroyAllWindows()
```
